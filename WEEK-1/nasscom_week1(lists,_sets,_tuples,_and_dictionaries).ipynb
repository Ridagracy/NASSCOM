```python
fruits = ['apple', 'banana', 'cherry', 'apple']
print("Original list:", fruits)

# Accessing elements by index
print("First fruit:", fruits[0])
```

```python
fruits.append('orange')
print("after appending:",fruits)
```

```python
fruits.remove('apple')
print("after removing:",fruits)
```

```python
print("number of fruits:",len(fruits))
```

```python
#2.tuple
dimensions=(10,20,30)
print("tuple:",dimensions)
```

```python
print("third element:",dimensions[2])
```

```python
unique_numbers={1,2,3,4,4,2}
print("set with no duplicates:",unique_numbers)
```

```python
unique_numbers.add(5)
print("set after adding 5:",unique_numbers)
```

```python
unique_numbers.remove(3)
print("set after removing:",unique_numbers)
```

```python
set_a={1,2,3}
set_b={3,4,5}
print("union:",set_a|set_b)
print("intersection:",set_a&set_b)
print("difference:",set_a-set_b)
```

```python
#dictionaries
student = {
    'name':'Rida',
    'age':17,
    'course':'Python'
}
print("Dictionary:", student)
```

```python
print("Student name:",student['name'])
```

```python
student['grade']='A+'
print("after adding grade:",student)
```

```python
del student['age']
print("after deleting age:",student)
```

```python
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
keys=['id','name','marks']
values=[538,'rida',97]
student_info=dict(zip(keys,values))
print("student info from zip:",student_info)
```
